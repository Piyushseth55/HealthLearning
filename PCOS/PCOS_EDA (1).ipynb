Here the column name "II    beta-HCG(mIU/mL)" have float values but its data type is object. So, we have to change its data type into float. In this column there is a value having string type which cannot be converted into float64. So, we create a function to convert the datatype if we found some ValueError then it return 'NaN' value.

```python
# Function to convert values to float

def to_float(value):
    try:
        return float(value)
    except ValueError:
        return np.nan

```

```python
# Change data type

pcos['II    beta-HCG(mIU/mL)'] = pcos['II    beta-HCG(mIU/mL)'].apply(to_float)
pcos['II    beta-HCG(mIU/mL)'] = pcos['II    beta-HCG(mIU/mL)'].astype('float64')

pcos['AMH(ng/mL)'] = pcos['AMH(ng/mL)'].apply(to_float)
pcos['AMH(ng/mL)'] = pcos['AMH(ng/mL)'].astype('float64')
```

```python
# calculate the skewness for each column

def calculate_skewness_loop(pcos):
  """
  Calculates skewness of each numeric column in the dataframe using a loop.

  Args:
      df: A pandas dataframe.

  Returns:
      A dictionary containing column names as keys and skewness values.
  """
  skewness = {}
  for col in pcos.columns:
    if not pd.api.types.is_numeric_dtype(pcos[col]):
      continue
    skewness[col] = pcos[col].skew()
  return skewness

result = calculate_skewness_loop(pcos.copy())

print("Skewness of each column:")
for col, skew in result.items():
    print(f"Column: {col}, Skewness: {skew:.2f}")  
```

# Forming a correlation matrix between the features and target.

```python
# correlation  matrix
plt.figure(figsize = (10,8))
sea.set(font_scale=0.6)
corr = pcos.corr()
sea.heatmap(corr)
```

```python
# correlation matrix
corr_matrix = pcos.corr().abs()
corr_matrix[target].sort_values()
```

```python
#calculate correlation from different methods


# Calculate Pearson's correlation
corr_pearson = pcos.corr()

# Calculate Spearman's rank correlation
corr_spearman = pcos.corr(method='spearman')

# Calculate Kendall's rank correlation
corr_kendall = pcos.corr(method='kendall')

plt.figure(figsize = (25,15))

# Print the different correlation matrices
print("Pearson's correlation:")
sea.heatmap(corr_pearson)

print("Spearman's rank correlation: ")
sea.heatmap(corr_spearman)

print("Kendall's rank correlation:")
sea.heatmap(corr_kendall)



```

# Visualizing the top 17 features with highest correlation with a correlation heatmap.

```python
plt.figure(figsize=(10,8))
sea.set(font_scale=0.6)
indexes = corr_matrix.nlargest(17, target)[target].index
cm = pd.DataFrame(np.corrcoef(pcos[indexes].values.T))
sea.heatmap(data = cm, annot = True, yticklabels = indexes.values,
            xticklabels = indexes.values, cbar =True, cmap = 'YlGnBu',
            square = True)
plt.title(f"correlation heatmap with {target}")
plt.show()
```

# Filtering out features with over 80% mutual correlation factor for identification.

```python
correlation_threshold = 0.8 # threshold = 80%
upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
highly_correlated_pairs = []

# Iterate through the DataFrame and find highly correlated feature pairs
for column in upper_triangle.columns:
    correlated_columns = upper_triangle.index[upper_triangle[column] >= correlation_threshold].tolist()
    for correlated_column in correlated_columns:
        pair = (column, correlated_column)
        highly_correlated_pairs.append(pair)
highly_correlated_pairs
```

# Detecting Outliers From Features
There are outliers present in the dataset. Distance -based models such as K-Nearest-Neighbours classifieres are particularly sensitive to outliers as they cluster the data into classes based on their proximity with other data points, 
and hence their presence can distort the model's results.
Also, models trained on datasets containing outliers may not generalize well to new, unseen data. 
Hence, removing those outliers is necessary for optimal performance.

```python
# box to visualize outliers
all_features = list(pcos.columns[1:])
features_df = pcos[all_features]
binary_features = features_df.columns[(features_df.max() == 1) & (features_df.min() == 0)]
nonbinary_data = features_df.drop(binary_features, axis=1)
for feature in nonbinary_data.columns.to_list():
  plt.figure(figsize=(5,3))
  sea.boxplot(data=pcos, x=target, y=feature, hue = target)
  plt.xlabel(target)
  plt.ylabel(feature)
  plt.title(f"Boxplot for {feature}")
  plt.show()
```
